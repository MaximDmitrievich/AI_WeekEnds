Напишем пару методов, которые облегчат нам вывод информации о распознавании изображения

````python
def resultOut(faces):
    for item in faces:
        print('FaceID: ' + str(item['faceId']))
        print('Emotions: ')
        print('    Contempt: ' + str(item['faceAttributes']['emotion']['contempt']))
        print('    Disgust: ' + str(item['faceAttributes']['emotion']['disgust']))
        print('    Fear: ' + str(item['faceAttributes']['emotion']['fear']))
        print('    Anger: ' + str(item['faceAttributes']['emotion']['anger']))
        print('    Happiness: ' + str(item['faceAttributes']['emotion']['happiness']))
        print('    Neutral: ' + str(item['faceAttributes']['emotion']['neutral']))
        print('    Sadness: ' + str(item['faceAttributes']['emotion']['sadness']))
        print('    Surprise: ' + str(item['faceAttributes']['emotion']['surprise']))
        print('Age: ' + str(item['faceAttributes']['age']))
        print('Gender: ' + str(item['faceAttributes']['gender']))
        print('\n\n')
````

#### Шаг 4. Распознавание.

Теперь, после того, как мы обучили модель, мы можем попробовать распознать лицо, которое присутствует у нас в репозитории.

Дополним метод вывода до такого образа

````python
def result_with_face(result, faces, names):
    for item in result:
        for face in faces:
            if len(face['candidates']) > 0 and face['faceId'] == item['faceId']:
                print('Name: ' + str(list(filter(lambda name: name['personId'] == face['candidates'][0]['personId'], names))[0]['name']))
        print('FaceID: ' + str(item['faceId']))
        print('Emotions: ')
        print('    Contempt: ' + str(item['faceAttributes']['emotion']['contempt']))
        print('    Disgust: ' + str(item['faceAttributes']['emotion']['disgust']))
        print('    Fear: ' + str(item['faceAttributes']['emotion']['fear']))
        print('    Anger: ' + str(item['faceAttributes']['emotion']['anger']))
        print('    Happiness: ' + str(item['faceAttributes']['emotion']['happiness']))
        print('    Neutral: ' + str(item['faceAttributes']['emotion']['neutral']))
        print('    Sadness: ' + str(item['faceAttributes']['emotion']['sadness']))
        print('    Surprise: ' + str(item['faceAttributes']['emotion']['surprise']))
        print('Age: ' + str(item['faceAttributes']['age']))
        print('Gender: ' + str(item['faceAttributes']['gender']))
        print('\n\n')
````

Теперь нам нужно вывести результат лиц с фото, результат лиц в репозитории (id) и имена в репозитории, которые сопоставлены с id пользователя и группы.

````python
result_faces = CF.face.detect(img_url, True, False, attributes)
names = CF.person.lists(person_group_id)
faceIds = [d['faceId'] for d in result_faces]
faces = CF.face.identify(faceIds, person_group_id=person_group_id)
result_with_face(result_faces, faces, names)
````
